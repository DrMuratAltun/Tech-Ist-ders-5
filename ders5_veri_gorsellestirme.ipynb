## Zaman Serisi İşlemlerinin Bir Bölümünü de yapıyoruz

```python
#Tarih veri türünü (formatını) kullanmak
```

```python
df['Date'].dtype #metin olarak giirrlmiş
```

```python
df['Date']=pd.to_datetime(df['Date'])#tarih formatına çevir
df['Date'].info() 
```

```python
df['day']=df['Date'].dt.day #haftanın kaçıncı günü
df['day']
```

```python
df['month']=df['Date'].dt.month # ay sırası numarası
df['month']
```

```python
df['day_name']=df['Date'].dt.day_name() #gün isimlerini al
df['day_name']
```

```python
df['month_name']=df['Date'].dt.month_name() # ay simlerini al
df['month_name']
```

```python
df['Time'].dtype #metin formatında
```

```python
df['Time']=pd.to_datetime(df['Time']) #tam tarih saat formatı
```

```python
df['Time']
```

```python
df['hour']=df['Time'].dt.hour # saat bilgisini alalım
df['hour']
```

```python
df['hour'].value_counts(normalize=True) # normalize=True oransal oalrak verir
```

```python
df['Gender'].value_counts(normalize=True) #normalize=True oransal oalrak verir
```

## Veri Görselleştirme

```python
#aylara göre alış veriş
plt.figure(figsize=(12,6)) #büyüklüğünü belirle
sns.countplot(x=df['month_name'])
#sns.countplot(data=df,x='month_name')
plt.title('Aylara göre alış veriş sayısı') #grafiğin başlığı
plt.xticks(rotation=45)#etiketlerin eğik biçimde yazılması
plt.show() #title filan eklnemesze gerek yok
```
